# Linear Regression

## Basic Setup

1. Start from training data $\left(\left(\boldsymbol{x}_{i}, y_{i}\right)\right)_{i=1}^{n}$, with $\boldsymbol{x}_{i} \in \mathbb{R}^{d}$ and $y_{i} \in \mathbb{R}$.
2. Model is a linear predictor: pick $\boldsymbol{w} \in \mathbb{R}^{d}$ with
$$
\boldsymbol{x}_{i} \mapsto \boldsymbol{w}^{\top} \boldsymbol{x}_{i}=: \hat{y}_{i} \approx y_{i} .
$$
3. Loss function $\ell$ is squared loss $\ell_{\text {sq }}$ (standard regression loss):
$$
\ell_{\mathrm{sq}}\left(\boldsymbol{w}^{\top} \boldsymbol{x}_{i}, y_{i}\right)=\frac{1}{2}\left(\boldsymbol{w}^{\top} \boldsymbol{x}_{i}-y_{i}\right)^{2} .
$$
We will minimize the empirical risk (average loss over training examples):
$$
\widehat{\mathcal{R}}(\boldsymbol{w})=\frac{1}{n} \sum_{i=1}^{n} \ell_{\mathrm{sq}}\left(\boldsymbol{w}^{\top} \boldsymbol{x}_{i}, y_{i}\right)=\frac{1}{2 n} \sum_{i=1}^{n}\left(\boldsymbol{w}^{\top} \boldsymbol{x}_{i}-y_{i}\right)^{2}=\frac{1}{2 n}\|\boldsymbol{X} \boldsymbol{w}-\boldsymbol{y}\|^{2} \quad \text { where } \boldsymbol{X}:=\left[\begin{array}{ccc}
\leftarrow & \boldsymbol{x}_{1}^{\top} & \rightarrow \\
& \vdots & \\
\leftarrow & \boldsymbol{x}_{n}^{\top} & \rightarrow
\end{array}\right] \in \mathbb{R}^{n \times d}
$$
